### **Dataset Overview**

* **Car full name**: The full name or designation of the vehicle, often combining make, model, and variant.
* **Make**: The brand or manufacturer of the car.
* **Model**: The specific model or version of the car.
* **Minimal price (gross) [PLN]**: The minimum retail price of the car, in Polish złoty (PLN).
* **Engine power [KM]**: The car's engine power, measured in horsepower (KM in Polish).
* **Maximum torque [Nm]**: The peak torque the engine can produce, measured in Newton-meters (Nm).
* **Type of brakes**: The braking system used, such as disc or drum brakes.
* **Drive type**: The drivetrain configuration, like FWD (front-wheel drive), RWD (rear-wheel drive), or AWD (all-wheel drive).
* **Battery capacity [kWh]**: Total energy capacity of the car’s battery, measured in kilowatt-hours (kWh).
* **Range (WLTP) [km]**: Estimated driving range on a full charge under WLTP standards, in kilometers.
* **Wheelbase [cm]**: The distance between the front and rear axles, in centimeters.
* **Length [cm]**: The overall length of the car, in centimeters.
* **Width [cm]**: The car’s width, in centimeters.
* **Height [cm]**: The car’s height, in centimeters.
* **Minimal empty weight [kg]**: The car’s minimum weight when empty, measured in kilograms.
* **Permissible gross weight [kg]**: Maximum legally allowed weight, including passengers and cargo, in kilograms.
* **Maximum load capacity [kg]**: The maximum weight the car can carry, in kilograms.
* **Number of seats**: The number of passenger seats in the car.
* **Number of doors**: The number of doors on the car.
* **Tire size [in]**: The tire size, measured in inches.
* **Maximum speed [kph]**: The top speed of the car, in kilometers per hour.
* **Boot capacity (VDA) [l]**: Trunk or cargo space capacity, measured in liters according to VDA standards.
* **Acceleration 0-100 kph [s]**: Time taken to accelerate from 0 to 100 kilometers per hour, in seconds.
* **Maximum DC charging power [kW]**: The highest charging power supported when using a DC fast charger, in kilowatts (kW).
* **Mean - Energy consumption [kWh/100 km]**: Average energy consumption per 100 kilometers, in kilowatt-hours (kWh).

### **Task 1: A customer has a budget of 350,000 PLN and wants an EV with a minimum range of 400 km.**

## **Task 2: Find the outliers in the mean - Energy consumption [kWh/100 km] column.**

```python
# 1. Calculate Quartiles (Q1 & Q3)
Q1 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.25)
Q3 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.75)

# 2. Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# 3. Define Bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 4. Find Outliers
outliers = df[(df['mean - Energy consumption [kWh/100 km]'] < lower_bound) | (df['mean - Energy consumption [kWh/100 km]'] > upper_bound)]
num_outliers = len(outliers)

# Print Mathematical Results
print(f"Q1 (25%): {Q1} kWh/100km")
print(f"Q3 (75%): {Q3} kWh/100km")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound:.2f}")
print(f"Upper Bound: {upper_bound:.2f}")
print(f"Min Value in Data: {df['mean - Energy consumption [kWh/100 km]'].min()}")
print(f"Max Value in Data: {df['mean - Energy consumption [kWh/100 km]'].max()}")
print(f"Total Outliers: {num_outliers}")
```

```python
fig = px.box(
    df, 
    y=df['mean - Energy consumption [kWh/100 km]'],
    title=f"Energy Consumption Box Plot Outliers Detected: {num_outliers}",
    points="all", # Show actual data points next to the box
    hover_data=["Make", "Model"], # Show car details on hover
)

fig.update_layout(
    yaxis_title="Energy Consumption (kWh/100 km)",
    xaxis_title="All Vehicles"
)

fig.show()
```

## **Task 4: Build an EV recommendation class.** 
    The class should allow users to input their budget, desired range, and battery capacity. The class should then return the top three EVs matching their criteria

```python
from IPython.display import display
class SimpleEVRecommender:
    """A basic class to filter EVs based on user criteria."""
    
    def __init__(self, data_frame):
        self.data = data_frame 

    def recommend(self, budget, min_range, min_capacity):
        
        # 1. FILTER: Find all matches
        matches = self.data[
            (self.data['Minimal price (gross) [PLN]'] <= budget) &
            (self.data['Range (WLTP) [km]'] >= min_range) &
            (self.data['Battery capacity [kWh]'] >= min_capacity)
        ]
        
        # 2. RANK: Sort by price (ascending = cheapest first)
        ranked = matches.sort_values(
            by='Minimal price (gross) [PLN]', 
            ascending=True
        )
        
        # 3. RETURN: Get the top 3 and select key columns
        return ranked.head(3)

# --- INTERACTIVE USAGE (Using input()) ---

if not df.empty:
    simple_recommender = SimpleEVRecommender(df)

    print("\nEV Recommendation")
    
    try:
        # Prompt user for input and convert immediately
        user_budget = float(input("Enter your MAXIMUM budget (PLN, e.g., 350000): "))
        user_min_range = float(input("Enter your MINIMUM required range (km, e.g., 400): "))
        user_min_capacity = float(input("Enter MINIMUM battery capacity (kWh, e.g., 60): "))

        # Run the recommendation logic
        test_results = simple_recommender.recommend(
            budget=user_budget, 
            min_range=user_min_range, 
            min_capacity=user_min_capacity
        )

        print("\nYour Top Recommendations")
        if test_results.empty:
            print("No electric vehicles found matching your criteria.")
        else:
            display(test_results)
            
    except ValueError:
        print("\nInput Error: Please ensure you enter valid numeric values for all inputs.")
```

### **Analysis of Task 4. EV Recommendation System analysis**

### 1. Objective
The goal was to create a Python Class that functions as a simple, automated tool to recommend the top three Electric Vehicles based on a user's **maximum budget**, **minimum range**, and **minimum battery capacity**.

### 2. Methodology
* **Structure:** a Python Class (`SimpleEVRecommender`) to make the filtering logic reusable.
* **Core Logic:** The `recommend` method uses straightforward **Boolean filtering** in Pandas to check if a car meets all three user conditions simultaneously.
* **Ranking:** The results are then simply sorted by **Minimal Price (Ascending)**, ensuring the cheapest suitable options are always presented first.
* **Output:** The system returns the top 3 best-matching cars found in the dataset.

### 3. Insight
This tool demonstrates the practical utility of data analysis by transforming static data into an **actionable recommendation engine**. It immediately identifies the most **cost-effective** and qualified long-range EVs for any customer profile, automating the initial purchase decision.

### **Task 5: Inferential Statistics – Hypothesis Testing: Test whether there is a significant** 
difference in the average Engine power [KM] of vehicles manufactured by two leading
manufacturers i.e. Tesla and Audi.
### 1. Objective
To test if there is a **statistically significant difference** in the average `Engine power [KM]` between vehicles manufactured by **Tesla** and **Audi**.

### 2. Hypotheses
* **Null Hypothesis ($H_0$):** The average engine power of Tesla is equal to Audi ($\mu_{Tesla} = \mu_{Audi}$).
* **Alternative Hypothesis ($H_1$):** The average engine power of Tesla is different from Audi ($\mu_{Tesla} \neq \mu_{Audi}$).

```python
from scipy import stats
# --- TASK 5: Hypothesis Testing ---

# 1. Extract the data for the two groups
tesla_power = df[df['Make'] == 'Tesla']['Engine power [KM]']
audi_power = df[df['Make'] == 'Audi']['Engine power [KM]']

# 2. Check Sample Means (for context)
print(f"Mean Engine Power (Tesla): {tesla_power.mean():.2f} KM")
print(f"Mean Engine Power (Audi):  {audi_power.mean():.2f} KM")
print("-" * 40)

# 3. Perform Two-Sample T-Test
# We use 'equal_var=False' (Welch's t-test) because we cannot assume 
# that both brands have the exact same variance in engine power.
t_stat, p_value = stats.ttest_ind(tesla_power, audi_power, equal_var=False)

# 4. Display Results
print("--- T-Test Results ---")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_value:.4f}")

# 5. Automated Conclusion
alpha = 0.05 # Significance level (5%)

if p_value < alpha:
    print("\nResult: REJECT the Null Hypothesis.")
    print("Conclusion: There IS a significant difference in engine power.")
else:
    print("\nResult: FAIL TO REJECT the Null Hypothesis.")
    print("Conclusion: There is NO significant difference in engine power.")
```

### **Analysis of Task 5: Hypothesis Testing Analysis**

## Findings
* **Tesla Mean Power:** 533.00 KM
* **Audi Mean Power:** 392.00 KM
* **P-value:** 0.1068

## Conclusion
Since the **P-value (0.1068)** is greater than the significance level (**0.05**), we **Fail to Reject the Null Hypothesis**.

**Insight:** Although the sample data shows Tesla has a higher average engine power, the difference is **not statistically significant**. This means we cannot rule out the possibility that this difference occurred purely by chance due to the small sample size. We cannot definitively claim that one brand produces more powerful engines based on this dataset alone.
